```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def analyticsExplorer(path, file):
    df = pd.read_excel(path)
    
    fig = plt.figure()
    ax1 = plt.subplot(1, 2, 1)
    ax1.set_title("Histograma")
    ax1.set_xlabel("conversion")
    df["conversion"].plot(kind="hist")
    ax2 = plt.subplot(1, 2, 2)
    ax2.set_title("Diagrama de caja")
    df["conversion"].plot(kind="box")
    fig.suptitle("Variables númericas del archivo {}".format(file))
    plt.show()
    
    fig = plt.figure()
    ax1 = plt.subplot(1, 4, 1)
    ax1.set_title("Campañas")
    table = df["Campaña"].value_counts()
    table.plot(kind="bar")
    ax2 = plt.subplot(1, 4, 2)
    ax2.set_title("Marcas")
    table = df["Marca"].value_counts()
    table.plot(kind="bar")
    ax3 = plt.subplot(1, 4, 3)
    ax3.set_title("Medios")
    table = df["Medio"].value_counts()
    table.plot(kind="bar")
    ax4 = plt.subplot(1, 4, 4)
    ax4.set_title("Fuentes")
    table = df["Fuente"].value_counts()
    table.plot(kind="bar")
    fig.suptitle("Variables categoricas del archivo {}".format(file))
    plt.show()
```

```python
def parrillaExplorer(path, file):
    df = pd.read_excel(path)
    
    fig = plt.figure()
    ax1 = plt.subplot(2, 2, 1)
    ax1.set_title("Histograma")
    ax1.set_xlabel("inversion")
    df["inversion"].plot(kind="hist")
    ax2 = plt.subplot(2, 2, 2)
    ax2.set_title("Diagrama de caja")
    df["inversion"].plot(kind="box")
    ax3 = plt.subplot(2, 2, 3)
    ax3.set_xlabel("rating")
    df["rating"].plot(kind="hist")
    ax4 = plt.subplot(2, 2, 4)
    df["rating"].plot(kind="box")
    fig.suptitle("Variables númericas del archivo {}".format(file))
    plt.show()
    
    fig = plt.figure()
    ax1 = plt.subplot(1, 4, 1)
    ax1.set_title("Canales")
    table = df["channel"].value_counts()
    table.plot(kind="bar")
    ax2 = plt.subplot(1, 4, 2)
    ax2.set_title("Formato")
    table = df["Formato"].value_counts()
    table.plot(kind="bar")
    ax3 = plt.subplot(1, 4, 3)
    ax3.set_title("Medios")
    table = df["medium"].value_counts()
    table.plot(kind="bar")
    ax4 = plt.subplot(1, 4, 4)
    ax4.set_title("Referencias")
    table = df["reference"].value_counts()
    table.plot(kind="bar")
    fig.suptitle("Variables categoricas del archivo {}".format(file))
    plt.show()
```

```python
def demografiaExplorer(path, file):
    df = pd.read_excel(path)
    
    fig = plt.figure()

    ax3 = plt.subplot(2, 2, 1)
    ax3.set_xlabel("Duración")
    df["Duración"].plot(kind="hist")
    ax4 = plt.subplot(2, 2, 2)
    df["Duración"].plot(kind="box")
    ax5 = plt.subplot(2, 2, 3)
    ax5.set_xlabel("Tiempo en página")
    df["Tiempo en página"].plot(kind="hist")
    ax6 = plt.subplot(2, 2, 4)
    df["Tiempo en página"].plot(kind="box")
    fig.suptitle("Variables númericas del archivo {}".format(file))
    plt.show()
    
    fig = plt.figure()
    ax1 = plt.subplot(1, 4, 1)
    ax1.set_title("Buscadores")
    table = df["browser"].value_counts()
    table.plot(kind="bar")
    ax2 = plt.subplot(1, 4, 2)
    ax2.set_title("Equipos")
    table = df["device"].value_counts()
    table.plot(kind="bar")
    ax3 = plt.subplot(1, 4, 3)
    ax3.set_title("Sistemas Operativos")
    table = df["Sistema Operativo"].value_counts()
    table.plot(kind="bar")
    ax4 = plt.subplot(1, 4, 4)
    ax4.set_title("Ciudades")
    table = df["Ciudad"].value_counts()
    table.plot(kind="bar")
    fig.suptitle("Variables categoricas del archivo {}".format(file))
    plt.show()
```

```python
def interactionsExplorer(path, file):
    df = pd.read_excel(path)
    
    fig = plt.figure()

    ax1 = plt.subplot(2, 2, 1)
    ax1.set_xlabel("Promedio de duración")
    df["Promedio de duración"].plot(kind="hist")
    ax2 = plt.subplot(2, 2, 2)
    df["Promedio de duración"].plot(kind="box")
    ax3 = plt.subplot(2, 2, 3)
    ax3.set_xlabel("Tiempo en página")
    df["Tiempo en página"].plot(kind="hist")
    ax4 = plt.subplot(2, 2, 4)
    df["Tiempo en página"].plot(kind="box")
    fig.suptitle("Variables númericas del archivo {}".format(file))
    plt.show()
    
    fig = plt.figure()
    plt.title("Hora del día")
    table = df["hour"].value_counts()
    table.plot(kind="bar")
    plt.show()
```

```python
def convertionsExplorer(path, file):
    df = pd.read_excel(path)
    
    fig = plt.figure()

    ax3 = plt.subplot(2, 2, 1)
    ax3.set_xlabel("Total de conversiones")
    df["Total de conversiones"].plot(kind="hist")
    ax4 = plt.subplot(2, 2, 2)
    df["Total de conversiones"].plot(kind="box")
    ax5 = plt.subplot(2, 2, 3)
    ax5.set_xlabel("Número objetivo de conversión")
    df["Número objetivo de conversión"].plot(kind="hist")
    ax6 = plt.subplot(2, 2, 4)
    df["Número objetivo de conversión"].plot(kind="box")
    fig.suptitle("Variables númericas del archivo {}".format(file))
    plt.show()
    
    fig = plt.figure()
    ax1 = plt.subplot(1, 3, 1)
    ax1.set_title("Campañas")
    table = df["Campaña"].value_counts()
    table.plot(kind="bar")
    ax2 = plt.subplot(1, 3, 2)
    ax2.set_title("Creatividad")
    table = df["Creatividad"].value_counts()
    table.plot(kind="bar")
    ax3 = plt.subplot(1, 3, 3)
    ax3.set_title("Madios")
    table = df["Medio"].value_counts()
    table.plot(kind="bar")
    fig.suptitle("Variables categoricas del archivo {}".format(file))
    plt.show()
```
